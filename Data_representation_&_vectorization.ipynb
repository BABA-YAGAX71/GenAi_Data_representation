#**One hot encoding**

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

#sample categorical data
data = np.array([['red'], ['blue'], ['green'], ['red'], ['blue']])

#Initialize the one hot encoder
encoder = OneHotEncoder(sparse_output=False)

#Fit and transform the data
encoded_data = encoder.fit_transform(data)

#Print the encoded data
print(encoded_data)
```

#**Bag Of Words(BOW)**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "text":["people watch ds with bappy",
            "ds with bappy watch ds with bappy",
            "people write comment",
            "ds with bappy write comment"],"output":[1,1,0,0]
})
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
bow = cv.fit_transform(df["text"])
```

```python
bow.toarray()
```

```python
#Vocabulary
print(cv.vocabulary_)
```

#N-Grams

```python
df = pd.DataFrame({
    "text":["people watch ds with bappy",
            "ds with bappy watch ds with bappy",
            "people write comment",
            "ds with bappy write comment"],"output":[1,1,0,0]
})
df
```

```python
#Bi grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
bow = cv.fit_transform(df["text"])
bow.toarray()
```

```python
import numpy as np
import pandas as pd
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
nltk.download('punk_tab')
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
story=[]
for filename in os.listdir('/content/Data/'):
  filepath = os.path.join('/content/Data', filename)
  if os.path.isfile(filepath):
    f = open(filepath)
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
      story.append(simple_preprocess(sent))
```

```python
story
```

```python
len(story)
```

```python
#Model

model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vec = model.wv.get_normed_vectors()
```
